```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
data_scaled = data.copy()

data_scaled['Education'] = label_encoder.fit_transform(data_scaled['Education'])
data_scaled['Marital_Status'] = label_encoder.fit_transform(data_scaled['Marital_Status'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
col_scale = ['Income', 'Kids', 'Expenses',
       'TotalAcceptedCmp', 'NumTotalPurchases', 'Customer_Age']

data_scaled[col_scale] = scaler.fit_transform(data_scaled[col_scale])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca.fit(data_scaled)
pca_data = pd.DataFrame(pca.transform(data_scaled), columns=(["PC_1","PC_2", "PC_3"]))
pca_data
```

```python
pca_data= pd.DataFrame(pca_data)

pca_data.to_csv('pca_data.csv', index=False)
pca_data
```

```python
x = pca_data['PC_1']
y = pca_data['PC_2']
z = pca_data['PC_3']
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(x, y, z, c='teal', marker='o')
ax.set_title("A 3D projection of our data after Dimensionality reduction")
plt.show()
```

```python
std_scaler= StandardScaler()
std_scaled= std_scaler.fit_transform(pca_data)
std_scaled
```

```python
#Hierarchical clustering

import scipy.cluster.hierarchy as sch
z= sch.linkage(std_scaled, method='complete', metric='euclidean')
z

plt.figure(figsize=(15,10))
plt.title('Hierarchical clustering dendrogram')
plt.xlabel('index')
plt.ylabel('distance')
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering


df= AgglomerativeClustering(n_clusters=7, linkage='complete', affinity='euclidean').fit(std_scaled)
cluster_labels= pd.Series(df.labels_)
cluster_labels
```

```python
pca_data['H_cluster']=cluster_labels
pca_data
```

```python
pca_data.H_cluster.value_counts()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(pca_data, pca_data['H_cluster'])
```

```python
fig = plt.figure(figsize=(12,12))
ax = plt.subplot(111, projection='3d', label="bla")

ax.scatter(x, y,  s=10, c=pca_data["H_cluster"], marker='o', cmap="viridis")
plt.show()
```

```python
#DBSCAN

from sklearn.cluster import DBSCAN
model= DBSCAN(eps=1.1, min_samples=3)
model.fit(std_scaled)
```

```python
Dcluster_labels= pd.Series(model.labels_)
Dcluster_labels
```

```python
pca_data['Dclusters']= Dcluster_labels
pca_data
```

```python
pca_data.Dclusters.value_counts()
```

```python
silhouette_score(pca_data, pca_data['Dclusters'])
```

```python
fig = plt.figure(figsize=(12,12))
ax = plt.subplot(111, projection='3d', label="bla")

ax.scatter(x, y,  s=10, c=pca_data["Dclusters"], marker='o', cmap="viridis")
plt.show()
```

```python
#KMeans

from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(std_scaled)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.axvline(x = 5, linestyle = '--', color = 'red')
plt.xlabel('K-Values')
plt.ylabel('WCSS')
plt.title('Elbow Graph')
```

```python
km = KMeans(n_clusters=5)
ypred = km.fit_predict(std_scaled)
ypred
```

```python
len(ypred)
```

```python
fig = plt.figure(figsize=(12,12))
ax = plt.subplot(111, projection='3d', label="bla")

ax.scatter(x, y, z, s=40, c=pca_data["Clusters"], marker='o', cmap="viridis")
ax.set_title("A 3D projection of our data divided into  5 clusters", fontsize = 14)
plt.show()
     
```
